Making prompts reusable

Separating logic from input

Allowing dynamic inputs

Instead of hardcoding prompts, we define a template and pass values later.

SECTION 3: Multi-Input Prompt System


In this section, we build a prompt that accepts multiple inputs instead of just one.

This makes prompts more flexible and useful in real-world applications.

We will use three inputs:

topic

audience

tone

This helps generate customized outputs for different users.

```python
from langchain_core.prompts import PromptTemplate

# Create multi-input template
template = "Explain {topic} for {audience} in a {tone} tone."

prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

# Test Case 1
output1 = prompt.format(
    topic="Artificial Intelligence",
    audience="beginners",
    tone="friendly"
)

print("Test Case 1:")
print(output1)

# Test Case 2
output2 = prompt.format(
    topic="Python",
    audience="kids",
    tone="fun"
)

print("\nTest Case 2:")
print(output2)

# Test Case 3
output3 = prompt.format(
    topic="Deep Learning",
    audience="engineers",
    tone="technical"
)

print("\nTest Case 3:")
print(output3)
```

SECTION 4: Prompt Variations Engine


In this section, we create multiple prompt templates for different use cases.

Instead of using one fixed template, we design different styles:

Teaching Style → step-by-step explanation

Interview Style → questions

Storytelling Style → creative explanation

This helps build flexible AI systems for different scenarios.

```python
from langchain_core.prompts import PromptTemplate

# Teaching Template
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step."
)

# Interview Template
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}."
)

# Storytelling Template
story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story."
)

# Input
topic = "Machine Learning"

# Outputs
print("Teaching Style:")
print(teaching_template.format(topic=topic))

print("\nInterview Style:")
print(interview_template.format(topic=topic))

print("\nStorytelling Style:")
print(story_template.format(topic=topic))
```

SECTION 5: ChatPromptTemplate (Roles System)


In this section, we use ChatPromptTemplate, which allows us to structure prompts using roles.

There are mainly 3 roles:

System → defines behavior of AI

Human → user input

AI → response

This is useful for building chatbots, assistants, and real-world AI applications.

```python
from langchain_core.prompts import ChatPromptTemplate

# Create chat template
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role}."),
    ("human", "Explain {topic} in simple terms.")
])

# Test Case 1
messages1 = chat_template.format_messages(
    role="teacher",
    topic="Neural Networks"
)

print("Test Case 1:")
for msg in messages1:
    print(msg)

# Test Case 2
messages2 = chat_template.format_messages(
    role="motivator",
    topic="Success"
)

print("\nTest Case 2:")
for msg in messages2:
    print(msg)
```

SECTION 6: Input Validation + Prompt Generator


In this section, we build a system that:

Validates user input

Generates structured prompts dynamically


We ensure:

Only valid values are accepted

Default values are assigned if needed


This is important in real-world applications to avoid errors and maintain consistency.

```python
from langchain_core.prompts import PromptTemplate

# Step 1: Validation Function
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience! Setting default: beginner")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone! Setting default: formal")
        tone = "formal"

    return audience, tone


# Step 2: Prompt Generator Function
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    # Different styles
    if style == "teaching":
        template = "Explain {topic} for {audience} in a {tone} teaching style."

    elif style == "interview":
        template = "Ask interview questions about {topic} for {audience} in a {tone} tone."

    elif style == "storytelling":
        template = "Explain {topic} as a story for {audience} in a {tone} tone."

    else:
        template = "Explain {topic} for {audience} in a {tone} tone."

    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    return prompt.format(
        topic=topic,
        audience=audience,
        tone=tone
    )


# 🔹 Test Case 1
print("Test Case 1:")
print(generate_prompt("AI", "beginner", "fun", "storytelling"))

# 🔹 Test Case 2 (Invalid input test)
print("\nTest Case 2:")
print(generate_prompt("Python", "kids", "cool", "teaching"))
```

SECTION 7: Template Reusability Test (Final Section)


In this section, we test reusability of a single prompt template.

Instead of creating multiple templates, we use one template and pass different inputs.

This proves:

The template is reusable

The system is scalable

Clean design is maintained


```python
from langchain_core.prompts import PromptTemplate

# One reusable template
template = "Explain {topic} for {audience} in a {tone} tone."

prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

# 5 different test cases
test_cases = [
    {"topic": "AI", "audience": "beginner", "tone": "fun"},
    {"topic": "Python", "audience": "intermediate", "tone": "casual"},
    {"topic": "Deep Learning", "audience": "expert", "tone": "formal"},
    {"topic": "Data Science", "audience": "beginner", "tone": "formal"},
    {"topic": "Machine Learning", "audience": "intermediate", "tone": "fun"},
]

# Run all test cases
for i, case in enumerate(test_cases, 1):
    print(f"Test Case {i}:")
    print(prompt.format(**case))
    print()
```
